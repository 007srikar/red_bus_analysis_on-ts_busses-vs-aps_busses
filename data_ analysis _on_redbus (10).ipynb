# **i have splitted the column named type to two names named bus_name and bus_type so that the unique values are reduced and the analysis will be easy and can be understand easily and droped the type column bellow**

```python
# Splitting the 'type' column into two new columns using "(" as the delimiter
c[['bus_name', 'bus_type']] = c['type'].str.split('(', expand=True, n=1)

# Handling cases where there's no '(' in the original 'type' column
c['bus_type'] = c['bus_type'].str.replace(')', '', regex=False).str.strip()
c['bus_name'] = c['bus_name'].str.strip()
```

```python
c.drop(columns=["type"],inplace=True)
```

# Variation in Ratings:

# Outliers:
+ The dots below or above the whiskers represent extreme ratings (very low or very high).
# Top Performers: 
+ Bus types with higher median lines and smaller variation (box) are better rated consistently.

```python
df_melted = c.melt(id_vars=['price'], 
                    value_vars=['Emergency', 'M_ticket', 'Charging_Point', 'LIVE_TRACKING'], 
                    var_name='Amenity', 
                    value_name='Available')


df_filtered = df_melted[df_melted['Available'] == 1]


plt.figure(figsize=(10, 6))
sns.boxplot(x='Amenity', y='price', data=df_filtered)

plt.title('Price Variation by Availability of Amenities')
plt.xlabel('Amenity')
plt.ylabel('Price')


plt.grid(True)
plt.show()
```

# Price Distribution Across Amenities:
+ The price distribution for buses offering amenities like "Emergency," "M_Ticket," "Charging_Point," and "LIVE_TRACKING" is generally similar, with most prices concentrated below ₹1000. However, outliers (higher-priced buses) exist for all amenities, particularly for "LIVE_TRACKING."
+ The interquartile range (IQR, the box) is narrow across all amenities, indicating that most buses with these features have similar pricing. However, buses with "Charging_Point" and "LIVE_TRACKING" show slightly higher price variability compared to others.
+ Outliers (dots above whiskers) indicate that a few buses with "LIVE_TRACKING" charge significantly more, suggesting premium pricing for real-time tracking.
+ Conclusion:
+ Buses with basic amenities like "Emergency" and "M_Ticket" are generally priced affordably and consistently. While "Charging_Point" and "LIVE_TRACKING" features do not guarantee higher average prices, they are associated with occasional premium-priced buses. This suggests that advanced amenities like "LIVE_TRACKING" might be the reasons for having  high-paying customers.
