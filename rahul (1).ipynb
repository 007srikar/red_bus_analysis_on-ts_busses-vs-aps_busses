```python
import requests 
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
import time 
import re
import numpy as np
import pandas as pd
```

```python
RTC=[]
ratings=[]
no_of_ratings=[]
type=[]
b_point=[]
d_point=[]
route=[]
fair_price=[]
Emergency=[]
M_ticket=[]
Charging_Point=[]
time_duration=[]
no_of_reviews=[]
LIVE_TRACKING=[]
text=""
driver=webdriver.Chrome()
a_link=["https://www.redbus.in/bus-tickets/hyderabad-to-vijayawada?fromCityId=124&toCityId=134&fromCityName=Hyderabad&toCityName=Vijayawada&busType=Any&srcCountry=IND&destCountry=null&onward=12-Nov-2024"]
for i in a_link:
    driver.get(i)
    time.sleep(5)
    for j in driver.find_elements(By.XPATH, "/html/body/section/div[2]/div[4]/div/div[2]/div/div[2]/div[2]/div[1]/div[1]/div[2]/div/div[4]/div[2]"):
        j.click()
        time.sleep(5)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(5)
    page=driver.page_source
    soup=BeautifulSoup(page)
    #text=text+soup.text
    for i in soup.find_all('div', class_="clearfix bus-item-details"):
          text = i.text
          print(i.text)
          print("*"*30)
          a = i.find("div",class_="bus-type f-12 m-top-16 l-color evBus")
          if a:
              type.append(a.text)
          #print("****"*20)
          c = i.find("div",class_="travels lh-24 f-bold d-color")
          if c:
                RTC.append(c.text)
          else:
                RTC.append(np.nan)
          d=i.find("div",class_="rating-sec lh-24")
          if d:
              ratings.append(d.text)
          else:
               ratings.append(np.nan)
          e=i.find("div",class_="filter-titles f-12 f-bold")
          if e:
              b_point.append(e.text)
          else:
                b_point.append(np.nan)
          f=i.find("div",class_="fare d-block")
          if f:
             fair_price.append(f.text)
          else:
                fair_price.append(np.nan)
          regex=re.findall("Emergency",text)
          if regex:
              Emergency.append(int(1))
          else:
              Emergency.append(int(0))
          regex=re.findall("M-ticket",text)
          if regex:
              M_ticket.append(int(1))
          else:
              M_ticket.append(int(0))
          regex=re.findall("Charging Point",text)
          if regex:
              Charging_Point.append(int(1))
          else:
             Charging_Point.append(int(0))
          regex=re.findall("[0-9]h.+[0-9]m+",text)
          if regex:
              time_duration.append(regex[0])
          else:
             time_duration.append(np.nan)
          g = i.find("div",class_="hero-feature fl clearfix")
          if g:
                LIVE_TRACKING.append(int(1))
          else:
                LIVE_TRACKING.append(int(0))
          h = i.find("div",class_="no-ppl l-color  m-top-16")
          if h:
                no_of_reviews.append(h.text)
          else:
                no_of_reviews.append(np.nan)
          k=i.find("div",class_="column-six p-right-10 w-10 fl")
          if k:
              no_of_ratings.append(k.text)
          else:
              no_of_ratings.append(np.nan)
          j=i.find("div",class_="bp-loc l-color w-wrap f-12 m-top-16")
          if j:
              d_point.append(j.text)
          else:
              d_point.append(np.nan)
         
```

```python
df1=pd.DataFrame({"RTC":RTC,"ratings":ratings,"type":type,"b_point":b_point,"fair_price":fair_price,"Emergency":Emergency,"M_ticket":M_ticket,"Charging_Point":Charging_Point,"time_duration":time_duration,"LIVE_TRACKING":LIVE_TRACKING,"No of Ratings":no_of_ratings,"Drop point":d_point})
```
